### **Select Evaluation Metric**

Model evaluation metric selection steps:
1. Find the class imbalance, 
    - No Class imbalance 
        - Use Accuracy
    - Class imbalance exists
        - **Ask yourself**
            - What is your objective?, 
            - what is the cost of FP or FN?
        - **Choose the metric**
            - You want to know how many positives are correct?
                - use Precision
            - You want to know how many actual positives were identified correctly?
                - Use Recall (e.g spam detection)
            - Both Precison and Recall are important 
                - Use F1 score
            

```python
print("0- Not Survived ,1 -Survived : Count")
print(df_RF_new['Survived'].value_counts())
print("0- Not Survived ,1 -Survived : Percentage")
print(df_RF_new['Survived'].value_counts(normalize=True)*100)
```

**Conclusion: Since there is a class imbalance, and the objective is to predict both survival and non survival, we can choose F1 Score**

### **GridSearch**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
rf = RandomForestClassifier()

cv_params = {'max_depth': [None],        #The maximum depth the tree will construct to before stopping
             'max_features': [8],        #The maximum of features selected for each base learner 
             'max_samples': [0.8],
             'min_samples_leaf': [1],    #The minimum number of samples that must be in each child node for the split to complete. 
             'min_samples_split': [6,8], #The minimum number of samples that a node must have to split into more nodes.
             'n_estimators': [10]        #The maximum number of base learners your ensemble will grow
             }

# 3. Define a dictionary of scoring metrics to capture
scoring = {'accuracy', 'precision', 'recall', 'f1'}

# 4. Instantiate the GridSearchCV object
rf_cv = GridSearchCV(rf, cv_params, scoring=scoring, cv=5, refit='f1')    
```

```python
import pickle 

# Define a path to the folder where you want to save the model
path = 'pikcle_files/'

def write_pickle(path, model_object, save_name:str):
    '''
    save_name is a string.
    '''
    with open(path + save_name + '.pickle', 'wb') as to_write:
        pickle.dump(model_object, to_write)
        

def read_pickle(path, saved_model_name:str):
    '''
    saved_model_name is a string.
    '''
    with open(path + saved_model_name + '.pickle', 'rb') as to_read:
        model = pickle.load(to_read)

        return model
    
#write_pickle(path,rf_cv,'automati_model')
rf_cv=read_pickle(path,"automati_model")
```
